# Feature engineering

```python
data['budget']=data['budget']/1000000
data['gross']=data['gross']/1000000
```

```python
data['profit']=data['gross']-data['budget']
```

```python
data[['movie_title','profit']].sort_values(by='profit',ascending=False).head(10)
```

```python
data.drop_duplicates(keep='first',inplace=True)
```

```python
data[['language']].value_counts()
```

```python
def Language(x):
        
        if x=="English":
            return "English"
        else:
            return 'Foreign'
```

```python
data['language']=data['language'].apply(Language)
```

```python
data['language'].value_counts()
```

```python
def duration(x):
    if x >120:
        return "Long movie"
    else:
        return "short movie"
```

```python
data['duration']=data['duration'].apply(duration)
```

```python
data['genres']
```

```python
data['genres'].value_counts()
```

```python
data['Moviegenra']=data['genres'].str.split('|')
data['genra1']=data["Moviegenra"].apply(lambda x:x[0])
data['genra2']=data["Moviegenra"].apply(lambda x:x[1] if len(x)>1 else x[0] )
data['genra3']=data["Moviegenra"].apply(lambda x:x[2] if len(x)>2 else x[0])
data['genra4']=data["Moviegenra"].apply(lambda x:x[3] if len(x)>3 else x[0])
```

```python
pd.set_option('max_rows',100)
@interact
def show_more_than(column='imdb_score',score=9.0):
    x=data[data[column]>score][['movie_title','language','imdb_score','profit','duration','gross']]
    x=x.drop_duplicates(keep='first')
    x=x.sort_values(by='imdb_score',ascending=True)
    x=x.style.background_gradient(cmap='autumn')
    
    return x
```

```python
pd.set_option('max_rows',100)
@interact
def show_more_than_article(column=['budget','gross'],x=1000):
    x=data[data[column]>x][['movie_title','language','imdb_score','budget','duration','gross']]
    x=x.style.background_gradient(cmap='autumn')
    
    return x
```

```python
pd.set_option('max_rows',100)
@interact_manual
def recommend_language(column=['Foreign','English']):
    y=data[data['language']==column][['movie_title','language','duration','imdb_score']].sort_values(by='imdb_score',ascending=False)
    y=y.style.background_gradient(cmap='winter')
    return y
```

```python
pd.set_option('max_rows',100)
@interact_manual
def recommend_MyHero_movies(column=data['actor_1_name'].unique().tolist()):
    y=data[data['actor_1_name']==column][['actor_1_name','movie_title','language','duration','imdb_score']].sort_values(by='imdb_score',ascending=False)
    y=y.style.background_gradient(cmap='winter')
    return y
```
